```python
#Ghadeer Amer.. Add fixed costs function to calculate property rent, insurance cost, and employee and manager costs.


def calculate_costs():
    x = input('Enter how you would to calculate cost: ')
    if(x == 'monthly'):
        print('The monthly cost is:',(rent/12 + insurance_cost/12 + employees_salary*employees+manger_salary),'SR')
    elif (x == 'annual'):
        print('The annual cost is:',(rent + insurance_cost + (employees_salary*employees*12)+(manger_salary*12)),'SR')
    elif (x == 'semi-annual'):
        print('The semi-annual cost is:',(rent/2 + insurance_cost/2 + (employees_salary*employees*6)+(manger_salary*6)),'SR')

calculate_costs()
```

```python
#Rawan Feaz .. Number of adults, children, and senior citizens in the theater should be randomized for each screen, theatre, and day. 

#This function make a random list of seats type
def randomize_seats():
    seats_type=['adult','childern','senior_citizens']
    seats=[]
    for i in range(1,31):
        seats.append(random.choice(seats_type))
    return seats
#This function will assign the random seats list from randomize_seats() to a screen
def randomize_screen():
    screen_seats=[]
    screen_seats.append(randomize_seats())
    return screen_seats
#This function will assign randomize screen seats from randomize_screen() for a Day
def randomized_day():
    day={'screen 1':randomize_screen(),
         'screen 2':randomize_screen(),
         'screen 3':randomize_screen(),
         'screen 4':randomize_screen(),
         'screen 5':randomize_screen()}
    return day
#This function will assign randomize screen seats for each day in week from randomized_day()
def screens_of_week():
    week={'Sunday':randomized_day(),
          'Monday':randomized_day(),
          'Tuesday':randomized_day(),
          'Wednesday':randomized_day(),
          'Thursday':randomized_day(),
          'Friday':randomized_day(),
          'Saturday':randomized_day()}
    return week
    
week=screens_of_week()
```
